```python
import pandas as pd
url = 'https://archives.nseindia.com/content/indices/ind_nifty500list.csv'
NSE500_data = pd.read_csv(url)
NSE500_symbols = NSE500_data['Symbol']
NSE500_symbols
```

```python
import requests
import bs4 as BeautifulSoup

from selenium import webdriver
from selenium.webdriver import ActionChains

from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager

import pandas as pd
import datetime
import time

Master_df = pd.DataFrame()

start_date = '15'
start_month = '06'
start_year = '2022'

end_date = '01'
end_month = '07'
end_year = '2022'

full_start_date = datetime.date(int(start_year),int(start_month),int(start_date))
full_end_date = datetime.date(int(end_year),int(end_month),int(end_date))

while full_start_date <= full_end_date:
    
    string_date = full_start_date.strftime('%d') + full_start_date.strftime('%m') + full_start_date.strftime('%Y')
    
    url = 'https://www1.nseindia.com/archives/equities/mto/MTO_{}.DAT'.format(string_date)
    
    driver = webdriver.Chrome(ChromeDriverManager().install())
    driver.get(url)
    
    Get_data = driver.find_element_by_tag_name('body')
    Get_data.text
    
    time.sleep(3)
    
    lines = Get_data.text.split("\n")
    
    if lines[0] == 'Not Found':
        full_start_date = full_start_date + datetime.timedelta(days=1)
        continue
        
    only_data = []
    
    for line in lines:
        cell_data = line.split(",")
        if cell_data[0] == '20':
            only_data.append(line)
            continue
        else:
            continue
            
    list_of_dictionaries = []
    
    for line in only_data:
        
        temp_dictionary = {}
        cell_data = line.split(",")
        
        temp_dictionary['RECORD TYPE'] = cell_data[0]
        temp_dictionary['SR NO'] = cell_data[1]
        temp_dictionary['NAME OF SECURITY'] = cell_data[2]
        temp_dictionary['SEGMENT'] = cell_data[3]
        temp_dictionary['QUANTITY TRADED'] = cell_data[4]
        temp_dictionary['DELIVERABLE QUANTITY'] = cell_data[5]
        temp_dictionary['% DELIVERABLE QUANTITY TO TRADED QUANTITY'] = cell_data[6]
        
        list_of_dictionaries.append(temp_dictionary)
        
    df = pd.DataFrame(list_of_dictionaries)
    df = df.drop(['RECORD TYPE','SR NO','QUANTITY TRADED','% DELIVERABLE QUANTITY TO TRADED QUANTITY'],axis = 1)
    df = df[df['SEGMENT']=='EQ']
    df = df.drop('SEGMENT',axis = 1)
    df = df[df['NAME OF SECURITY'].isin(NSE500_symbols)]
    df = df.set_index("NAME OF SECURITY")
    df = df.transpose()
    df.insert(loc = 0, column = 'DATE', value = full_start_date)
    
    Master_df = pd.concat([Master_df, df], ignore_index=True)
    
    driver.close()
    
    full_start_date = full_start_date + datetime.timedelta(days=1)
    
Master_df
```

```python
col_names = list(Master_df.columns.values)

Final_df = pd.DataFrame()

for symbol in col_names:
    list_of_dicts = []
    i = 0
    while i in range(0,len(Master_df['{}'.format(symbol)])):
        if pd.isna(Master_df['{}'.format(symbol)][i]) == True:
            i+=1
            continue
        else:
            temp_dict = {}
            temp_dict['{} DATE'.format(symbol)] = Master_df['DATE'][i]
            temp_dict['{} DEL QTY'.format(symbol)] = Master_df['{}'.format(symbol)][i]
            list_of_dicts.append(temp_dict)
            i+=1        
    df = pd.DataFrame(list_of_dicts)
    Final_df = pd.concat([Final_df,df],axis = 1)
    
Final_df
```
